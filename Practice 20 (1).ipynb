```python
import time

start = time.time()
print("Start time:", start)

mat = [1, 2, 3]
mat2 = [4, 5, 5]
add = []

for i, k in enumerate(mat):
    addition = mat[i] + mat2[i]
    add.append(addition)

print("Result:", add)

end = time.time()
print("End time:", end)

total_time = end - start
print("Total time to execute this program is:", total_time)
```

```python
# Vectorization
import time
import numpy as np
start=time.time()
mat=[1,2,3]
mat2=[4,5,5]
mat=np.array(mat)
mat2=np.array(mat2)
add= mat + mat2
print(add)
end=time.time()
print("Total time to execute this program is: ",(start-end))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data from uniform distribution
uniform_data = np.random.uniform(low=0, high=1, size=1000)

# Plot the histogram
plt.hist(uniform_data, bins=30, edgecolor='black')
plt.title('Histogram of Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```
